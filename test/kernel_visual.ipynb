```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.cm as cm
```

```python
# generated kde data
generated_data = np.genfromtxt('gauss_kernel_trace.csv')

# evaluated truth
def gaussian_kernel(x):
    return np.exp(-x*x/2.0) * 1 / np.sqrt(2*np.pi)
x = np.linspace(-5, 5, 1000)
evaluated_gauss_kernel = np.apply_along_axis(gaussian_kernel, 0, x)
```

```python
fig = plt.figure(figsize=(7*1.618,7))
ax = fig.add_subplot(1,1,1)
ax.plot(generated_data[:,0], generated_data[:,1],'+')
ax.plot(x, evaluated_gauss_kernel,'-')
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'$density$', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_title(r'$K(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}$', fontsize=20, y=1.08);
```
