```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
arr_cpu = np.genfromtxt('test_kde11_cpu.csv')
arr_gpu = np.genfromtxt('test_kde11_gpu.csv')

fig = plt.figure(figsize=(5.5*1.618, 5.5))
ax = fig.add_subplot(1,1,1); ax.set_yscale('log');

ax.plot(arr_cpu[:,0], arr_cpu[:,1], 'o', markersize=8, markeredgewidth=2,
        color='None', markeredgecolor='deepskyblue', label=r'CPU direct')
ax.plot(arr_cpu[:,0], arr_cpu[:,2], 'o', markersize=8, markeredgewidth=2,
        color='deepskyblue', markeredgecolor='deepskyblue', label=r'CPU dual tree')
ax.plot(arr_gpu[:,0], arr_gpu[:,1], 'o', markersize=8, markeredgewidth=2,
        color='None', markeredgecolor='limegreen', label=r'GPU direct')
ax.plot(arr_gpu[:,0], arr_gpu[:,2], 'o', markersize=8, markeredgewidth=2,
        color='limegreen', markeredgecolor='limegreen', label=r'GPU dual tree')

ax.set_title('running time', fontsize=20)
ax.set_xlabel(r'$\log N$', fontsize=20)
ax.set_ylabel(r'milli seconds', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=16)

ax.legend(loc=4, numpoints=1);
```
