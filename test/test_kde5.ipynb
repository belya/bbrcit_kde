```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.genfromtxt('saved/test_kde5_1Em5.csv')
fig = plt.figure(figsize=(5*1.618, 5))
ax = fig.add_subplot(1,1,1); ax.set_yscale('log');
ax.plot(arr[:,0], arr[:,1], 'k+', label='build time');
ax.plot(arr[0:6,0], arr[0:6,3], 'ro', label='naive');
ax.plot(arr[:,0], arr[:,4], 'g.-', label='single tree, epan');
ax.plot(arr[:,0], arr[:,5], 'b.-', label='dual tree, epan');
ax.plot(arr[:,0], arr[:,8], 'g.--', label='single tree, gauss');
ax.plot(arr[:,0], arr[:,9], 'b.--', label='dual tree, gauss');
ax.set_title('running time', fontsize=20)
ax.set_xlabel(r'$\log N$', fontsize=20)
ax.set_ylabel(r'$\log$(seconds)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.legend(loc=4, numpoints=1);
```
