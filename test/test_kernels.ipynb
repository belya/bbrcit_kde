```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from kde_visual_utils import plot_contour2d
```

```python
arr = np.genfromtxt('test_epanechnikov1d.csv')
fig = plt.figure(figsize=(5*1.618, 5))
ax = fig.add_subplot(1,1,1)
ax.plot(arr[:,0], arr[:,1], 'r', label='h = 1.0');
ax.plot(arr[:,0], arr[:,2], 'g', label='h = 0.5');
ax.plot(arr[:,0], arr[:,3], 'b', label='h=2.0');
ax.set_title('Epanechnikov');
ax.set_xlim(-2,2);
ax.legend();
```

```python
fig = plt.figure(figsize=(24,30))
ax = fig.add_subplot(3,2,1)
plot_contour2d('test_epanechnikov2d.csv', ax=ax, title=r'Epanechnikov', nlevels=5)
ax.set_xlim(-2,2);
ax.set_ylim(-2,2);
```

```python
arr = np.genfromtxt('test_gauss1d.csv')
fig = plt.figure(figsize=(5*1.618, 5))
ax = fig.add_subplot(1,1,1)
ax.plot(arr[:,0], arr[:,1], 'r', label='h = 1.0');
ax.plot(arr[:,0], arr[:,2], 'g', label='h = 0.5');
ax.plot(arr[:,0], arr[:,3], 'b', label='h=2.0');
ax.set_title('Gaussian');
ax.set_xlim(-4,4);
ax.legend();
```

```python
fig = plt.figure(figsize=(24,30))
ax = fig.add_subplot(3,2,1)
plot_contour2d('test_gauss2d.csv', ax=ax, title=r'Gaussian', nlevels=5)
ax.set_xlim(-5,5);
ax.set_ylim(-5,5);
```
