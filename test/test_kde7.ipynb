```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.genfromtxt('saved/test_kde7_dual.csv')
arr2 = np.genfromtxt('saved/test_kde7_single.csv')
arr3 = np.genfromtxt('saved/test_kde7_naive.csv')

fig = plt.figure(figsize=(5*1.618, 5))
ax = fig.add_subplot(1,1,1); ax.set_yscale('log');

ax.plot(arr3[:,0], arr3[:,6], 'ko', label=r'naive')
ax.plot(arr1[:,0], arr1[:,1], 'bs', label=r'dual ($h=10^{-1}$)')
ax.plot(arr2[:,0], arr2[:,1], 'rs', label=r'single ($h=10^{-1}$)')
ax.plot(arr1[:,0], arr1[:,2], 'b^', label=r'dual ($h=10^{-2}$)')
ax.plot(arr2[:,0], arr2[:,2], 'r^', label=r'single ($h=10^{-2}$)')

ax.set_title('running time', fontsize=20)
ax.set_xlabel(r'$\log N$', fontsize=20)
ax.set_ylabel(r'$\log$(seconds)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=16)

ax.legend(loc=4, numpoints=1);
```
