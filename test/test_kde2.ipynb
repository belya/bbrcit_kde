```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr = np.genfromtxt('test_kde2.csv')
fig = plt.figure(figsize=(7*1.618, 7))
ax = fig.add_subplot(2,2,1)
ax.plot(arr[:,0], arr[:,1]);
ax.set_title('naive');
ax = fig.add_subplot(2,2,2)
ax.plot(arr[:,0], arr[:,2]);
ax.set_title('single tree');
ax = fig.add_subplot(2,2,3)
ax.plot(arr[:,0], arr[:,3]);
ax.set_title('dual tree');
```
