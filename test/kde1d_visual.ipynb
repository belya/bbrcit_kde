```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.cm as cm
```

```python
generated_data = np.genfromtxt('bimodal_gauss1d.csv')
kde_data = np.genfromtxt('kde1d_gauss1d.csv')
print generated_data.shape, kde_data.shape
```

```python
fig = plt.figure(figsize=(5*1.618, 5))
ax = fig.add_subplot(1,1,1)
ax.hist(generated_data, 100, histtype='stepfilled', alpha=0.3, normed=True)
ax.plot(kde_data[:,0], kde_data[:,2], '-')
ax.set_xlabel(r'$X_1$', fontsize=20)
ax.set_ylabel(r'$Counts$', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_title('Input Sample', fontsize=20);
```
