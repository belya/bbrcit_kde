```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from kde_visual_utils import plot_data2d, plot_contour2d
```

```python
fig = plt.figure(figsize=(24,24))

ax = fig.add_subplot(2,2,1)
plot_data2d('example_scatter.csv', ax=ax, title=r'input sample', xlim=[-3.0,2.0], ylim=[-1.5,2.5])
ax = fig.add_subplot(2,2,3)
plot_contour2d('example_kde.csv', ax=ax, title=r'cross validated', 
               nlevels=10, colorbar=False, contour_fmt='%1.2f')
ax = fig.add_subplot(2,2,4)
plot_contour2d('example_akde.csv', ax=ax, title=r'cross validated and adapted', 
               nlevels=10, colorbar=False, contour_fmt='%1.2f')
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1,1,1)
plot_contour2d('example_gridsearch.csv', ax=ax, title=r'least squares cross validation score', 
               nlevels=30, nlevels_f=200, colorbar=True, cmap=plt.cm.Reds, contour_fmt='%1.4f')
```
